```python
# must specify path to data directory
DATA_DIR = '/home/peter/citadel/edgarScraper/data'

def read_in_data():

    dfs = []
    for year in range(1994, 2019):
        fileName = 'results_{}.csv'.format(year)
        filePath = os.path.join(DATA_DIR, fileName)
        dfs.append(pd.read_csv(filePath))

    df = pd.concat(dfs)
    df.drop('Unnamed: 0', axis=1, inplace=True)
    df['DATE'] = pd.to_datetime(df['DATE'])
    df.drop_duplicates(inplace=True)
    #df.set_index(['DATE', 'CIK', 'NAME'], inplace=True)
    #df.sort_index(inplace=True)
    return df

df = read_in_data()
```

```python

def getAggregatePerformanceStats(df):
    countsByQuarter = df.groupby(
        [pd.Grouper(key='DATE', freq='Y'), 'EXTRACTCODE']
    )['FINALSHORTTERM', 'FINALLONGTERM', 'CIK'].count().unstack()

    total = countsByQuarter['CIK'].sum(axis=1)
    total.name = "Total Count"
    totalShortTerm = countsByQuarter['FINALSHORTTERM'].sum(axis=1).div(total)
    totalShortTerm.name = "Short Term (%)"
    totalLongTerm = countsByQuarter['FINALLONGTERM'].sum(axis=1).div(total)
    totalLongTerm.name = "Long Term (%)"
    totalBadUrls = countsByQuarter['CIK']['BAD_URL'].div(total)
    totalBadUrls.name = "Bad Url (%)"
    totalFails = countsByQuarter['CIK']['FAILED'].div(total)
    totalFails.name = "Failure (%)"
    results = pd.concat([totalShortTerm, totalLongTerm, totalBadUrls, totalFails], axis=1).fillna(0)
    return results

def getStatsForExtractCode(df):
    countsByQuarter = df.groupby(
        pd.Grouper(key='DATE', freq='Y')
    )['FINALSHORTTERM', 'FINALLONGTERM', 'CIK'].count()
    total = countsByQuarter['CIK']
    totalShortTerm = countsByQuarter['FINALSHORTTERM'].div(total)
    totalShortTerm.name = "Short Term (%)"  
    totalLongTerm = countsByQuarter['FINALLONGTERM'].div(total)
    totalLongTerm.name = "Long Term (%)"
    results = pd.concat([totalShortTerm, totalLongTerm], axis=1).fillna(0)
    return results

def makePerformanceChart(df):
    fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(2,2,figsize=(16,12))
    aggResults = getAggregatePerformanceStats(df)
    textResults = getStatsForExtractCode(df[df['EXTRACTCODE'] == 'TEXT'])
    htmlResults = getStatsForExtractCode(df[df['EXTRACTCODE'] == 'HTML'])
    xbrlResults = getStatsForExtractCode(df[df['EXTRACTCODE'] == 'XBRL'])
    ax1 = aggResults.plot(ax=ax1, marker='x')
    ax1.set_title('Aggregate')
    ax1.legend(loc='lower right')
    ax1.grid()

    ax2 = textResults.plot(ax=ax2, marker='x')
    ax2.set_title('Text Only')
    ax2.legend(loc='lower right')
    ax2.grid()

    ax3 = htmlResults.plot(ax=ax3, marker='x')
    ax3.set_title('Html Only')
    ax3.legend(loc='lower right')
    ax3.grid()

    ax4 = xbrlResults.plot(ax=ax4, marker='x')
    ax4.set_title('Xbrl Only')
    ax4.legend(loc='lower right')
    ax4.grid()
    
    ylims = (0,1)
    ax1.set_ylim(ylims)
    ax2.set_ylim(ylims)
    ax3.set_ylim(ylims)
    ax4.set_ylim(ylims)
    
    ax2.set_xlim(ax1.get_xlim())
    ax3.set_xlim(ax1.get_xlim())
    ax4.set_xlim(ax1.get_xlim())
    
    x_label = ax1.axes.get_xaxis().get_label()
    x_label.set_visible(False)
    
    x_label = ax2.axes.get_xaxis().get_label()
    x_label.set_visible(False)
    
    fig.suptitle('Performance Metrics', fontsize=14)
    

makePerformanceChart(df)
```

```python
import seaborn as sns

def preparePlotData(df):
    _df = df.copy()
    _df[['FINALSHORTTERM', 'FINALLONGTERM']] = _df[['FINALSHORTTERM', 'FINALLONGTERM']].clip(lower=0, upper=1e12)
    debtByQuarter = _df.groupby(
        pd.Grouper(key='DATE', freq='Y')
    )['FINALSHORTTERM', 'FINALLONGTERM'].agg({'mean':'mean', 'std':'std'})
    
    longTermMean = debtByQuarter['mean']['FINALLONGTERM']
    longTermStd = debtByQuarter['std']['FINALLONGTERM']
    longTermMean.name = "longTermMean"
    longTermHigh = longTermMean + longTermStd
    longTermHigh.name = "longTermMean + std"
    longTermLow = (longTermMean - longTermStd).clip(lower=0)
    longTermLow.name = "longTermMean - std"
    
    shortTermMean = debtByQuarter['mean']['FINALSHORTTERM']
    shortTermStd = debtByQuarter['std']['FINALSHORTTERM']
    shortTermMean.name = "shortTermMean"
    shortTermHigh = shortTermMean + shortTermStd
    shortTermHigh.name = "shortTermMean + std"
    shortTermLow = (shortTermMean - shortTermStd).clip(lower=0)
    shortTermLow.name = "shortTermMean - std"    
    
    results = pd.concat(
        [longTermMean, longTermHigh, longTermLow, shortTermMean, shortTermHigh, shortTermLow],
        axis=1
    ).reset_index()

    return results

def setupAxes(ax1, ax2, data):

    l1 = ax1.plot_date(data['DATE'], data['longTermMean'], '-')
    ax1.set_yscale("log", nonposy='clip')

    d = data['DATE'].values
    f1 = ax1.fill_between(d, data['longTermMean + std'], data['longTermMean - std'],
                    facecolor='yellow', alpha=0.2)

    l2 = ax2.plot_date(data['DATE'], data['shortTermMean'], '-')
    ax2.set_yscale("log", nonposy='clip')
    d = data['DATE'].values
    f2 = ax2.fill_between(d, data['shortTermMean + std'], data['shortTermMean - std'],
                    facecolor='yellow', alpha=0.2)
    
    return (ax1, ax2, l1, l2, f1, f2)

def makeDebtPlot(df):
    fig, ((ax1, ax2),(ax3, ax4),(ax5, ax6), (ax7, ax8)) = plt.subplots(4,2,figsize=(16,16))
    aggResults = preparePlotData(df)
    ax1, ax2, l1, l2, f1, f2 = setupAxes(ax1, ax2, aggResults)
    textResults = preparePlotData(df[df['EXTRACTCODE']=='TEXT'])
    ax3, ax4, l3, l4, f3, f4 = setupAxes(ax3, ax4, textResults)
    htmlResults = preparePlotData(df[df['EXTRACTCODE']=='HTML'])
    ax5, ax6, l6, l5, f5, f6 = setupAxes(ax5, ax6, htmlResults)
    xbrlResults = preparePlotData(df[df['EXTRACTCODE']=='XBRL'])
    ax7, ax8, l7, l8, f7, f8 = setupAxes(ax7, ax8, xbrlResults)
    
    xlims = ax1.get_xlim()
    ylimsLongTerm = (1e5, 1e11)
    ylimsShortTerm = (1e5, 1e11)

    for i,ax in enumerate([ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8]):
        ax.set_xlim(xlims)
        if i % 2 == 0:
            ax.set_ylim(ylimsLongTerm)
        else:
            ax.set_ylim(ylimsShortTerm)
            
    ax1.set_title('Long Term Debt (Aggregate)')
    ax1.legend([(l1[0], f1), ], ['Avg. +/- Std.'], loc='upper left')
    ax1.grid()

    ax2.set_title('Short Term Debt (Aggregate)')
    ax2.legend([(l2[0], f2), ], ['Avg. +/- Std.'], loc='upper left')
    ax2.grid()

    ax3.set_title('Long Term Debt (Text Only)')
    ax3.legend([(l3[0], f3), ], ['Avg. +/- Std.'], loc='upper left')
    ax3.grid()
    
    ax4.set_title('Short Term Debt (Text Only)')
    ax4.legend([(l4[0], f4), ], ['Avg. +/- Std.'], loc='upper left')
    ax4.grid()
    
    ax5.set_title('Long Term Debt (HTML Only)')
    ax5.legend([(l5[0], f5), ], ['Avg. +/- Std.'], loc='upper left')
    ax5.grid()
    
    ax6.set_title('Short Term Debt (HTML Only)')
    ax6.legend([(l6[0], f6), ], ['Avg. +/- Std.'], loc='upper left')
    ax6.grid()
    
    ax7.set_title('Long Term Debt (XRBL Only)')
    ax7.legend([(l7[0], f7), ], ['Avg. +/- Std.'], loc='upper left')
    ax7.grid()
    
    ax8.set_title('Short Term Debt (XRBL Only)')
    ax8.legend([(l8[0], f8), ], ['Avg. +/- Std.'], loc='upper left')
    ax8.grid()
    
    fig.suptitle('Debt Over Time (log scaled)', fontsize=14)


makeDebtPlot(df)
```
