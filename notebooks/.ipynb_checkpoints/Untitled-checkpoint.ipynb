```python
# must specify path to data directory
DATA_DIR = '/home/peter/citadel/edgarScraper/data'

def read_in_data():

    dfs = []
    for year in range(1994, 2019):
        fileName = 'results_{}.csv'.format(year)
        filePath = os.path.join(DATA_DIR, fileName)
        dfs.append(pd.read_csv(filePath))

    df = pd.concat(dfs)
    df.drop('Unnamed: 0', axis=1, inplace=True)
    df['DATE'] = pd.to_datetime(df['DATE'])
    df.drop_duplicates(inplace=True)
    #df.set_index(['DATE', 'CIK', 'NAME'], inplace=True)
    #df.sort_index(inplace=True)
    return df

df = read_in_data()
```

```python
df.loc[
    (
        df.EXTRACTCODE.isin(['TEXT', 'HTML'])
    ),
    'FINALLONGTERM'
] = df.loc[
    (
        df.EXTRACTCODE.isin(['TEXT', 'HTML'])
    ), 'FINALLONGTERM'
]*1000 
```

```python
df.groupby(pd.Grouper(key='DATE', freq='Q'))['FINALLONGTERM'].median().plot()
```
